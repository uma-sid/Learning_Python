```python
iris_df = sns.load_dataset('iris')
```

# 2. 2-D Scatter plot

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris_df)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris_df, hue='species')
```

```python
sns.FacetGrid(data=iris_df, hue='species', size=6).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
```

Observations:
1. The sepal length and sepal width can be used to separate setosa from the other 2 categories

2. We can not separate the versicolor and virginica using sepal length and sepal width

```python
sns.set_style('whitegrid')
sns.FacetGrid(data=iris_df, hue='species', size=6).map(plt.scatter,'petal_length','petal_width').add_legend()
```

Observations:

1. Petal length and Petal width can also be used to separate the Setosa flowers from others.

2. Petal length and petal width can separate the versicolor and verginica to some extent.

# 3. Pair Plot

```python
plt.figure(figsize=(8,4))
pg = sns.PairGrid(data=iris_df, hue='species')
pg.map_lower(plt.scatter, marker='.')
pg.map_diag(plt.hist)
pg.map_upper(sns.kdeplot)
```

Observations:
    
    1. Petal length and petal widths each alone can separate our setosa flowers much more easier than the sepal length and sepal width together.
