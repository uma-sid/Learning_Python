## 6.3 PairGrid

Pair Grid is similar to pairplot but with more control and so many customizations.

```python
iris_df = sns.load_dataset('iris')
# create empty grid system based on iris dataset
iris_g = sns.PairGrid(iris_df, hue='species')
iris_g.map(plt.scatter)
```

```python
# Customizations
# putting different plots across the diagonal and scatter everywhere else

iris_df = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris_df, hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter) # this is to avoid scatter on the diagonal and plot scatter everywhere else.
```

```python
# creating scatter plot on the upper half, kde plots on the lower half and histograms along the diagonal
iris_df = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris_df, hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
# Creating custom grid instead of the deafault grid, which means defining x & y variables ourselves. In this we are
# avoiding all the available possible comparisions among the variables and customizing to look at only required 
# column comparisions.

iris_df = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris_df, hue='species', x_vars=['sepal_length', 'sepal_width'],
                                              y_vars=['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()

```

## 6.4 Facet Grid

Facet Grid gives us a lot of miniplots and with lot of variations. This is mainly useful when we have quite a few categories in the dataset. We can create plots of distribution of variables(histogram), scatter plots (btw 2 variables) based on each category of each variable.

We will use tips dataset as it has many categorical variables like sex, smoker, day and time.

```python
tips_df.head()
```

```python
# Here we are first creating the grid. We pass the categories of variables as rows and columns.
g = sns.FacetGrid(tips_df, col='day', row='time')
g
```

```python
# we are looking at the distribution plots of the total_bill column but at different scenarios like when it is 
# thursday and sex = female etc.
tips_g = sns.FacetGrid(tips_df, col='day', row='sex')
tips_g.map(sns.histplot, 'total_bill')
```

```python
tips_g = sns.FacetGrid(tips_df, col='day', row='sex')
tips_g.map(plt.scatter, 'total_bill', 'tip')
```

```python
# height and aspect ratio are used to adjust the size of the plot.
# col_order=[] is used to customize the order of the plots.
tips_g = sns.FacetGrid(tips_df, col='day', row='time', hue='smoker', height=4, aspect=0.7,
                          col_order=['Sun','Sat','Fri','Thur'], palette='Set1')
tips_g.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```

Note: In the grid creation we can use col_wrap=1/2/3, option to limit the no.of plots in a row.

```python
kws = dict(s=50, linewidth=0.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3, hue_order=['Yes', 'No'],
                           hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```

```python
att_df = sns.load_dataset('attention')
att_df.head()
```
