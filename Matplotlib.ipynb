# Matplotlib.pyplot

<b>Matplotlib.pyplot</b> is a collection of command style functions that make matplotlib work like MATLAB

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("Values")   # label for y-axis
plt.xlabel('Indices')  # label for x-axis
plt.title('My Plot')   # Main title for the whole plot
plt.show()
```

When you give only list of values for plottting then, the <b>indices of that list of values are taken as x-axis values by default.</b>

 When you pass 2 lists the first list is x-axis and 2nd list is considered as corresponding y-axis values.

```python
plt.plot([1,2,3,4],[2,4,6,7])
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Sample Plot 2')
plt.grid()     # grid set on to show the grid

plt.show()
```

The default plot is a line plot, but if you want a scatter plot instead, put an extra parameter in the plt.plot() at the end

## Scatter Plots

```python
plt.plot([1,2,3,4], [3,5,7,2], 'ro')   # r-means red colour and o-means 0-shaped dots
plt.grid()
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)

plt.plot(t, t**2,   'b--', label='^2')     # b-means blue, -- means dashed points, label is used for legends
plt.plot(t, t**2.25, 'rs', label='^2.25')  # r-means red, s -- means squares
plt.plot(t, t**2.5, 'g^', label='^2.5')    # g-means green, ^--means triangles
plt.grid()
plt.legend()    # add legend
plt.show()
```

## Controlling Line Properties

```python
x = [1,2,3,4]
y = [2,4,6,1]

plt.plot(x,y, linewidth=5.5) # linewidth is used to control the line width.
plt.grid()
```

### Using setp (Set Properties)

```python
x1 = [1,2,3,4]
y1 = [2,4,6,8]
x2 = [1,2,3,4]
y2 = [1,4,9,16]

lines = plt.plot(x1,y1,x2,y2) # with this command we are creating 2 line objects x1,y1 and x2,y2 and storing them 
                              # in a lines object. 
    
# Now we can control the properties of the 2 lines properties individually.
# using keyword args
plt.setp(lines[0], color='r', linewidth=2.0)
#or
# using MATLAB style string value pairs
plt.setp(lines[1], 'color', 'g', 'linewidth', '2.5')  # We can use any of these 2 styles for setting properties

plt.grid()
    
    
```

# Working with multiple figures and axes

```python
import numpy as np
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0,0.1)
t2 = np.arange(0.0, 5.0,0.02)

plt.figure(1)  # this creates the figure 1... We might create many subplots in it.

plt.subplot(211)  # 211 means, we are telling the matplot lib that there are 2 rows and 1 column of subplots in it 
                  # and we are plotting the first plot among those
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)  # 212 means, we are telling the matplot lib that there are 2 rows and 1 column of subplots in it 
                  # and we are plotting the 2nd plot among those
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)   # creating figure-1/creating space for figure-1
plt.subplot(211) # dividing the figure-1 in to 2 subplots one below the other and plotting the first one
plt.plot([1,2,3])
plt.subplot(212) # plotting the 2nd subplot
plt.plot([4,5,6])

plt.figure(2)   # creating a figure-2
plt.plot([5,6,7,2]) # Plotting the figure-2 directing without creating any subplots under it.

# Now, if we want to go back to fig-1's subplot-1 and do some changes, then do in the following way
plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1,2,3')
plt.show()
```
