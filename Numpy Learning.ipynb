# 4. Copies and Views

```python
a = np.arange(10)
a
```

```python
b = a[::2]
b
```

```python
np.shares_memory(a,b) # Internally for memory optimization, numpy basically b is pointing to the same memory location 
                      #using something called views.
```

```python
b[0] = 10
b
```

```python
np.shares_memory(a,b)   # as the memory is shared and we changed the b the zeroth element of a is also changed
```

```python
a
```

```python
# If you want to copy without sharing memory then use .copy() function.

a = np.arange(10)
c = a[::2].copy() # force a copy
c
```

```python
np.shares_memory(a,c)
```

```python
c[0] = 10
a
```

# 5. Fancy Indexing

NumPy arrays can be indexed with slices, but also with boolean or integer arrays(masks). This method is called <b>fancy indexing</b>. It creates copies not views. 

#### Using Boolean Mask

```python
a = np.random.randint(1,20,15)
a
```

```python
mask = (a%2 == 0)
mask
```

```python
even_numbers = a[mask]  # here with this operation there is a new copy created for even_numbers varible 
                        # and views technique is not used internally.
even_numbers
```

```python
a[mask] = -1
a
```

## 7.3 Broadcasting

 Basic operations on numpy arrays (addition, etc.) are elementwise

This works on arrays of the same size. Nevertheless, Its also possible to do operations on arrays of different sizes if NumPy can transform these arrays so that they all have the same size: this conversion is called broadcasting.

The image below gives an example of broadcasting:

```python
a = np.arange(0,40,10)
a
```

```python
a = np.tile(a, (3,1))  # the vector is broadcasted to replicate the vector row-wise 3 times and keep the vector
                        # as it is column wise as the number is only 1.
a
```

```python
a = a.T
a
```

```python
b = np.tile(np.arange(3), (4,1))
b
```

```python
a+b
```

```python
a = np.arange(0,40,10)
a.shape
```

```python
a = a[:, np.newaxis] # adds a new axis --> 2D array
a.shape
```

```python
a
```

```python
a+b
```
